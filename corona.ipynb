```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
df=pd.read_csv('train.csv')
df2=pd.DataFrame()
```

```python
df2=df
df3=df2.drop(["Id","Province/State","Lat","Long","Date"],axis=1)
df4= df3[["Country/Region","ConfirmedCases"]]
df3.iplot(kind='bar',x='Country/Region',y='Fatalities')
```

```python
india = df.loc[df['Country/Region'] == 'India']
india_covid=india[["Date","ConfirmedCases","Fatalities"]]
india_covid.iplot(kind='line',x="Date",y="ConfirmedCases",title="Covid-19 Cases in India")
india_covid.iplot(kind='line',x="Date",y="Fatalities",title="Covid-19 Deaths in India")
```
